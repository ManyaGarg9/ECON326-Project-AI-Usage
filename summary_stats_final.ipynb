```r
library(tidyverse)
library(haven)
library(dplyr)
library(scales)
library(tidyr)
```

```r
my_data <- read.csv("censusdata.csv")
head(my_data)
```

<h2>Summary statsitics </h2>

```r
my_data <- my_data |> select(NOC21)
head(my_data)
```

```r
my_data <- my_data |> filter(NOC21 != 99 & NOC21!= 88, na.rm = TRUE)
head(my_data)
     
```

```r

total_obs <- my_data |> nrow()
total_obs
```

In total, the dataset contains 28,782 observations, providing a large sample size for summarizing patterns across occupations and industries. 

```r
summary_stats <- my_data |> group_by(NOC21) |> 
                 summarise(count = n())
#proportion = count/total_obs, 
                           #std_dev = proportion*(1-proportion)/count)
summary_stats

stats_table <- summary_stats |> summarise(mean = mean(count), standard_deviation = sd(count),
                                         max = max(count))
stats_table
```

There are 26 distinct NOC21 occupation groups represented in the data, with counts ranging from 182 to 2,988 observations per group.

On average, there are 1,107 observations per NOC21 group with a standard deviation of about 708, suggesting substantial variation in the size of different occupation categories

```r
ai_data <- read.csv("AIusage.csv") |> rename(C1 = Business.characteristics)
head(ai_data)
```

```r
ai_data <- ai_data %>% select(C1, VALUE) |>
  filter(C1 %in% c('Agriculture, forestry, fishing and hunting [11]',
                   'Mining, quarrying, and oil and gas extraction [21]',
                   'Construction [23]',
                   'Manufacturing [31-33]',
                   'Wholesale trade [41]',
                   'Retail trade [44-45]',
                   'Transportation and warehousing [48-49]',
                   'Information and cultural industries [51]',
                   'Finance and insurance [52]',
                   'Real estate and rental and leasing [53]',
                   'Professional, scientific and technical services [54]',
                   'Administrative and support, waste management and remediation services [56]',
                   'Health care and social assistance [62]',
                   'Arts, entertainment and recreation [71]',
                   'Accommodation and food services [72]',
                   'Other services (except public administration) [81]'))
ai_data

stats_table <- summarise(ai_data, mean = mean(VALUE), sd = sd(VALUE), max = max(VALUE))
stats_table
```

The highest reported AI usage is <b>35.6%</b>, showing that at least one sector has a relatively high level of AI adoption compared with the overall average of <b>12.7%<b>.

Professional, scientific and technical services <b>(31.7%)</b> and health care and social assistance <b>(17.4%)</b> are among the sectors with above-average AI adoption, whereas accommodation and food services <b>(1.5%)</b> and ‘other services’ <b>(7.3%)</b> lag behind the overall mean.

Overall, the summary statistics show large differences in both the number of observations by occupation (NOC21) and the level of AI adoption by industry, highlighting substantial heterogeneity in how AI is used across the economy.
